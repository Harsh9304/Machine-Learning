 CORRELATION

 1.POSITIVE CORRELATION
 2.NEGATIVE CORRELATION

 13 columns means 13 variable and how these variables are correlated to each other is shown using correlation
 means in boston price predicting dataset
 price and crime are negatively correlated with each other as if crime increases then price decreases. But no of rooms and price are positively correlated as if no. of rooms inc. then price also inc.


```python
import pandas as pd
import numpy as np

data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
boston_dataset = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target = raw_df.values[1::2, 2]
```

```python
# pandas dataframe
# Define the feature names manually as the loaded data is a numpy array
feature_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'PRICE']
boston_df = pd.DataFrame(boston_dataset, columns=feature_names[:-1]) # Create DataFrame with original features
boston_df['PRICE'] = target # Add the 'PRICE' column
```

```python
boston_df.corr()
```
