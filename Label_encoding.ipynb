Label Encoding

Converting the labels into the numeric form

```python
#importing the Dependencies
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

Label Encoding  of Breast Cancer Dataset


```python
#loading the data from csv file to pandas dataframe
cancer_data=pd.read_csv('/content/breast_cancer_data.csv')
```

```python
cancer_data.head() # in diagnosis column we have two things M and B
'''M → Malignant

The tumor is cancerous and can spread to other parts of the body.

Requires further treatment.

B → Benign

The tumor is non-cancerous and usually does not spread.

Less dangerous, but still monitored.'''
```

```python
#finding the count of different labels
cancer_data['diagnosis'].value_counts()
```

```python
#load the Label encoder function
label_encode=LabelEncoder() # create a object of the Labelencoder
```

```python
labels=label_encode.fit_transform(cancer_data.diagnosis)
```

```python
#appending the labels to theh dataframe
cancer_data['target']=labels
```

```python
cancer_data.head()
```

as we havent mentioned the

0-->Benign

1-->Malignant

it will place alphabetically B for 0 and M for 1
